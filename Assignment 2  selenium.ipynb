# Q8: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:

```python
import selenium
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
dr=webdriver.Chrome(r'C:\Users\Admin\Downloads\chromedriver_win32.zip.exe')
dr.get('https://www.azquotes.com')
```

```python
tq=dr.find_element(By.XPATH,'//a[@href="/top_quotes.html"]')
tq.click()
```

```python
quotes=[]
author=[]
types_of_quotes=[]


q=dr.find_elements(By.XPATH,'//a[@class="title"]')
for i in q[0:1000]:
    qo=i.text
    quotes.append(qo)
    
```

```python
a=dr.find_elements(By.XPATH,'//div[@class="author"]')
for i in a[0:1000]:
    au=i.text
    author.append(au)
```

```python
tq=dr.find_elements(By.XPATH,'//div[@class="tags"]')
for i in tq[0:1000]:
    tqo=i.text
    types_of_quotes.append(tqo)
```

```python
st=0
end=10

for d in range(st,end):
    
    qa=dr.find_elements(By.XPATH,'//a[@class="title"]')
    for i in qa[0:1000]:
        qo=i.text
        quotes.append(qo)
    ab=dr.find_elements(By.XPATH,'//div[@class="author"]')
    for i in ab[0:1000]:
        au=i.text
        author.append(au)
    tqa=dr.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in tqa[0:1000]:
        tqo=i.text
        types_of_quotes.append(tqo)
   
```

```python
df=pd.DataFrame({'quotes':quotes,'types of quotes':types_of_quotes,'author':author})
df.head(1000)
```
